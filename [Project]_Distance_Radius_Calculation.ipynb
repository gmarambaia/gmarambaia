```python
acai_seed = pd.read_csv(r'C:\Users\Gabriel\Downloads\drive-download-20230107T151220Z-001\Buyers - Acai Seed.csv')
acai_seed.dropna(inplace = True)

acai_seed.head()
```

```python
acai_seed.info()
```

```python
# creating a tuple column (Latitude, Longitude)
acai_seed['coordinates'] = list(zip(acai_seed.Latitude, acai_seed.Longitude))

acai_seed.head()
```

```python
# defining a  function to calculate distance between two locations 
# loc1= location of an existing seller
# loc2= location of buyer

def distance_from(loc1,loc2): 
    dist=hs.haversine(loc1,loc2)
    return round(dist,2)
```

```python
# running a loop which will parse customers location one by one to distance from function 
for _,row in buyers.iterrows():
    acai_seed[row.Index]=acai_seed['coordinates'].apply(lambda x: distance_from(row.coordinates,x))
```

```python
acai_seed.head()
```

```python
acai_seed = acai_seed.set_index('Address')
acai_seed.head()
```

```python
acai_seed = acai_seed.T
acai_seed.head()
```

```python
acai_seed = acai_seed.drop(labels=['Latitude','Longitude','coordinates'], axis=0)
acai_seed.head()
```

```python
acai_seed = acai_seed.reset_index(level=0)
acai_seed.head()
```

```python
acai_seed = acai_seed.melt(id_vars=['index'], var_name=['Address'], value_name="Value")

acai_seed.head()
```
